Building a Model 
Using OLS [Ordinary Least Square Method
OLS it involves modelling the relationship between a dependent variable and one or more independent variable. OLS is used for find the best fit line by minimising the residuals.

```python
df.columns
```

```python
# importing libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols  # ordinary least squares
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import statsmodels.formula.api as smf
```

```python
# keep predicted with target variable
# spliting the data for train and test
trn=df.iloc[0:439,]
tst=df.iloc[439:,]
```

```python
df.columns
```

```python
# building the model
from statsmodels.formula.api import ols
import statsmodels.formula.api as smf

model1=smf.ols('Selling_Price ~ Brand+Model+Year+Seller_Type+Owner+KM_Driven+Ex_Showroom_Price',data=tst).fit()
print(model1.summary())
```

```python
tst.info()
```

```python
x_tst=tst.iloc[:,tst.columns!='Selling_Price']   # only the predictor data
y_tst=tst.iloc[:,tst.columns=='Selling_Price']   # only the target variable
```

```python
# pr2ediction on test data
pred_1=model1.predict(x_tst)
```

```python
# residuals
# if you want to find the residuals we want to convert target variable from dataframe to series
y_tst_series=y_tst.iloc[:,0]
```

```python
# now you can perform the residuals 
residuals=y_tst_series-pred_1
residuals
```

```python
# rmse
# finding the square of error
se_1=residuals.pow(2)
# finding the sum of it
sse_1=se_1.sum()
# finding the mean of it
mse=sse_1/len(se_1)
# finding the square root 
import math
rmse=math.sqrt(mse)
rmse
```

```python
len(y_tst_series)
```

```python
import matplotlib.pyplot as plt
obsno=np.arange(0,187,1)
fig,ax=plt.subplots()
ax.plot(obsno,y_tst_series,'o',label='Data')
ax.plot(obsno,pred_1,'r-',label='Predicted')
ax.legend(loc='best')
```
